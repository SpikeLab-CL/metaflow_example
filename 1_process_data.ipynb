# ¿Por qué nos gusta metaflow?

Lo vemos como una buena herramienta para las personas que les gusta trabajar en jupyter notebook y, sobre todo, aquellos que desarrollan un flujo o pipeline en varios jupyter notebooks.

+ Está todo en un script de `python`: no es necesario aprender un lenguaje de configuración nuevo
+ Funciona bien en una máquina: no es necesario levantar o configurar un cluster
+ Fácil para hacer copy/paste de jupyter notebooks.


## Limitaciones que vemos (sobre todo comparado con Airflow y Luigi)

+ No hay interfaz gráfica / panel de control siempre online
+ no es tan bueno para muchos procesos corriendo de forma frecuente
+ no es tan bueno para muchos usuarios al mismo tiempo
+ No es tan bueno el control de qué pipelines se han ejecutado 
+ Limitado a Python (*)


```python
import pandas as pd
import numpy as np
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", 
                 sep=';')
df.head(4)
```

```python
from sklearn.model_selection import (train_test_split, GridSearchCV,
                                    RandomizedSearchCV, cross_val_score)
from sklearn import model_selection
from sklearn import preprocessing
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, mean_absolute_error

bins = (2, 6, 8)
group_names = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins = bins, labels = group_names)
label_quality = preprocessing.LabelEncoder()
# Bad becomes 0 and good becomes 1
df['quality'] = label_quality.fit_transform(df['quality'])
df['quality'].value_counts()

X = df.drop('quality', axis=1)
y = df['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=50)
sc = preprocessing.StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)


#Save to disk
np.save('X_train.npy', X_train)
np.save('y_train.npy', y_train)

np.save('X_test.npy', X_test)
np.save('y_test.npy', y_test)
```
